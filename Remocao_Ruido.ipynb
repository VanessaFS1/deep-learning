```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread('orqu√≠dea.jpg')
plt.imshow(img)
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
```

```python
plt.imshow(dst)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dst),plt.title('Media 1x')
plt.show()
plt.clf()
```

```python
blur = cv.blur(img,(5,5))
plt.imshow(blur)
```
