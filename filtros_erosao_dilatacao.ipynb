```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('assinaturateste.png',0)

struct = np.ones((5,5), np.uint8)
```

```python
img_erosao = cv2.erode(img, struct)
plt.imshow(img_erosao)
```

```python
img_dilatacao = cv2.dilate(img, struct)
plt.imshow(img_dilatacao)
```

```python
img_erosao2 = cv.erode(img_dilatacao, struct, iterations=3)
plt.imshow(img_erosao2)
```
